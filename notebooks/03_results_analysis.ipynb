# Results Analysis

Analyze model predictions and evaluation results.

```python
import sys
sys.path.append('..')

import json
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load evaluation results
with open('../outputs/reports/evaluation_results.json') as f:
    results = json.load(f)

print('Evaluation Results:')
for k, v in results.items():
    print(f'  {k}: {v:.4f}')
```

```python
# Precision/Recall at different K values
k_values = [1, 3, 5, 10]
precision_at_k = [results.get(f'precision_at_{k}', 0) for k in k_values]
recall_at_k = [results.get(f'recall_at_{k}', 0) for k in k_values]

plt.figure(figsize=(8, 5))
plt.plot(k_values, precision_at_k, 'o-', label='Precision@K')
plt.plot(k_values, recall_at_k, 's-', label='Recall@K')
plt.xlabel('K')
plt.ylabel('Score')
plt.legend()
plt.title('Precision and Recall at K')
plt.grid(alpha=0.3)
plt.show()
```

```python
# Load predictions
with open('../outputs/predictions/predictions.json') as f:
    predictions = json.load(f)

print(f'Loaded {len(predictions)} predictions')
for p in predictions[:3]:
    print(f"\n{p['description']}")
    for idx, prob in p['predictions'][:3]:
        print(f"  Chemical {idx}: {prob:.2%}")
```
