# Smell-to-Molecule: Data Exploration

This notebook explores the smell-molecule dataset.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import json
from collections import Counter
```

```python
# Load data
df = pd.read_csv('../data/processed/train.csv')
print(f'Loaded {len(df)} samples')
df.head()
```

```python
# Description length distribution
df['desc_length'] = df['description'].apply(len)
plt.figure(figsize=(10, 4))
plt.hist(df['desc_length'], bins=50)
plt.xlabel('Description Length')
plt.ylabel('Count')
plt.title('Distribution of Description Lengths')
plt.show()
```

```python
# Chemical frequency analysis
all_chemicals = []
for chem_str in df['chemicals']:
    chemicals = json.loads(chem_str)
    all_chemicals.extend(chemicals)

chem_counts = Counter(all_chemicals)
top_20 = chem_counts.most_common(20)
print('Top 20 chemicals:', top_20)
```

```python
# Plot top chemicals
names = [c[0] for c in top_20]
counts = [c[1] for c in top_20]
plt.figure(figsize=(12, 6))
plt.barh(range(len(names)), counts)
plt.yticks(range(len(names)), names)
plt.xlabel('Frequency')
plt.title('Top 20 Chemicals by Frequency')
plt.tight_layout()
plt.show()
```
